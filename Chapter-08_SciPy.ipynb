```python
# Save figures as files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

chpt = 'Chpt08'
def save_fig(fig, figure_name):
    file_path = f'/content/drive/MyDrive/[WIP]Kennedy-Noah-Python-DataScience-Book/images-2021-06-05/Colored/{chpt}/{figure_name}.png'
    print(os.path.exists(file_path))
    fig.savefig(file_path, dpi=1000, format='png')
    convert_to_greyscale(file_path, figure_name)


def convert_to_greyscale(file_path, figure_name):   
    image = Image.open(file_path).convert("L")
    arr = np.asarray(image)
    target_path = f'/content/drive/MyDrive/[WIP]Kennedy-Noah-Python-DataScience-Book/images-2021-06-05/{chpt}/{figure_name}.svg'
    
    
    plt.imsave(target_path, arr, cmap='gray', vmin=0, vmax=255, format='svg')
```

Continuous Distributions
all have location(L), and scale(S) parameters and various shape parameters.

```python
N = stats.norm()
rvs = N.rvs(size=100000)
rvs
```

```python
plt.hist(rvs, bins=1000)
save_fig(plt, 'Figure.08.05')
plt.show()
```

```python
# For norm location is mean, scale is standard deviation
N1 = stats.norm(loc=30,scale=50)
rvs = N1.rvs(size=100000)
plt.hist(rvs, bins=1000)
save_fig(plt, 'Figure.08.06')
plt.show()
```

```python
import numpy as np
data = np.random.normal(0, 0.5, 1000)

mean, var  = stats.distributions.norm.fit(data)

x = np.linspace(-5,5,100)

fitted_data = stats.distributions.norm.pdf(x, mean, var)

fig = plt.figure()
plt.hist(data, density=True)
#save_fig(plt, 'Figure.09.10')
plt.plot(x, fitted_data,'r-')
```
